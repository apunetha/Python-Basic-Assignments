```python
def values_of_dict(d):
    """function to return list of all values in dict"""
    ans=[]
    for i in d.values():
        if type(i)==dict:
            ans2= values_of_dict(i)
            ans = ans +ans2
            continue
        if type(i) == list:
            ans =ans+i
            continue
        if type(i) ==tuple:
            for j in i:
                ans.append(j)
            continue
        if type(i)==set:
            ans=ans+list(i)
            continue
        ans.append(i)
    return ans
print(values_of_dict({'a':1,'b':2,'c':{'d':4,'e':5},'f':6,'g':[1,2,3],'h':(3,4,5),'i':{1,2,3},'j':"aman",'k':True,'l':34+7j}))
```

```python
def concate_list(*args):
    """Function containing n number of lists and 
    concate the elements of list into one list and returning """
    ans=[]
    for i in args:
        ans=ans+i
    return ans
concate_list([1,2,3],[1,2,3],[1,2,3],[1,2,3])
```

```python
def indexes_in_list(l):
    """return the dict of all the elements in list having value as indexes"""
    dic ={}
    for i in range(len(l)):
        if l[i] in dic:
            dic[l[i]].append(i)
        else:
            dic[l[i]] =[i]
    return dic
indexes_in_list([1,2,3,3,4,5,3,2,5])
```

```python
import PyPDF2

def PDFmerge(pdf1,pdf2,output):
    #creating pdf file merger object
    pdfMerger = PyPDF2.PdfFileMerger()
    
    #appending pdfs in object
    pdfMerger.append(pdf1)
    pdfMerger.append(pdf2)
    
    #writing combined pdf to output pdf file
    with open(output,'wb') as f:
        pdfMerger.write(f)
#pdf1
pdf1 =r"C:\Users\AMAN\Downloads\spider man.pdf"
#pdf2
pdf2=r"C:\Users\AMAN\Downloads\Booking Confirmation.pdf"
#result merger pdf
output = "combined_example.pdf"
#calling function to merge pdf
PDFmerge(pdf1,pdf2,output)
```
